```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
btcDf = pd.read_csv('../Data/crypto/bitcoinAll_1HR.csv')
dogeDf = pd.read_csv('../Data/crypto/dogecoinAll_1HR.csv')
```

```python
splitDate = btcDf[["time_period_start"]]
splitDate = np.array(splitDate)
date = []
time = []
for i in splitDate:
    datei = i[0]
    datei = datei.split("T")
    date.append(datei[0])
    time.append(datei[1])
    

btcDf["Date"] = date
data = btcDf["Date"] 
month = []
day = []
year = []
for i in data:
    datei = i.split("-")
    year.append(int(datei[0]))
    month.append(int(datei[1]))
    day.append(int(datei[2]))
    
btcDf['hour']=[int(d.split('T')[1].split(':')[0]) for d in btcDf['time_period_start']]
btcDf["year"] = year
btcDf["month"] = month
btcDf["day"] = day
btcDf = btcDf.drop(columns=["time_period_start", "time_period_end","time_open","time_close"])
print("BTC")
btcDf
```

```python
splitDate = dogeDf[["time_period_start"]]
splitDate = np.array(splitDate)
date = []
time = []
for i in splitDate:
    datei = i[0]
    datei = datei.split("T")
    date.append(datei[0])
    time.append(datei[1])
    

dogeDf["Date"] = date
data = dogeDf["Date"] 
month = []
day = []
year = []
for i in data:
    datei = i.split("-")
    year.append(int(datei[0]))
    month.append(int(datei[1]))
    day.append(int(datei[2]))
    
dogeDf['hour']=[int(d.split('T')[1].split(':')[0]) for d in dogeDf['time_period_start']]
dogeDf["year"] = year
dogeDf["month"] = month
dogeDf["day"] = day
dogeDf = dogeDf.drop(columns=["time_period_start", "time_period_end","time_open","time_close"])

print("DOGE")
dogeDf
```
